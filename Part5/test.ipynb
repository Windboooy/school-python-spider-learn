## 5.3 正则表达式

```python
title='你好,hello,世界'
pattern=re.compile(r'[\u4e00-\u9fa5]+')
result=pattern.findall(title)
print(result)
```

```python
def displaymatch(match):
    if match is None:
        return None
    return'<Match: %r, groups=%r>' % (match.group(), match.groups())
```

```python
valid = re.compile(r"^[a2-9tjqk]{5}$")
displaymatch(valid.match("akt5q"))
```

```python
displaymatch(valid.match("akt5e"))
```

```python
displaymatch(valid.match("akt"))
```

```python
displaymatch(valid.match("727ak"))
```

```python
pair = re.compile(r"^.*(.).*\1")
displaymatch(pair.match("717ak"))
```

```python
displaymatch(pair.match("718ak"))
```

```python
#创建root节点，并为其添加属性
root = etree.Element("root", intersting = "totally")
print(etree.tostring(root))
```

```python
#再次给root节点添加age属性
root.set("age", "30")
print(etree.tostring(root))
```

```python
root.text = "I am root"
print(etree.tostring(root))
print(root.text)
```

### 5.6.7获取拉勾网城市信息

```python
url = "https://www.lagou.com/lbs/getAllCitySearchLabels.json"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
html = response.read()
print(html)
```

```python
jsonobj = json.loads(html)
city_list = jsonpath.jsonpath(jsonobj,"$..name")
print(city_list)
```

```python
file = open('city.json','w')
content = json.dumps(city_list,ensure_ascii=False)
print(content)
file.write(content)
file.close()
```
